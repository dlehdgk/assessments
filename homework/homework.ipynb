```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#p3 differentiation hw
diff = pd.read_csv('data/p3_diff.csv')
#p3 integration hw
inte = pd.read_csv('data/p3_int.csv')
#p3 numerical methods hw
numerical = pd.read_csv('data/numerical.csv')
#p3 past paper hw
pp = pd.read_csv('data/p3_hw.csv')
#p4 contradiction hw
contra = pd.read_csv('data/contradiction.csv')
```

```python
df = diff[diff['Name']!='Total']
data = f'mean: %.3f, standard deviation: %.3f'%(np.mean(df['Mark']),np.std(df['Mark']))
hist = sns.histplot(df, x = 'Mark',kde=True)
hist.set_title('P3 Differentiation Homework')
hist.annotate(data, (0.1,4.9))
plt.savefig('figures/p3_ch6.png',dpi=300)
```
