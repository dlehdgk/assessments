```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data/implicit_diff.csv')
hw = df[df['Name']!='Total']
```

```python
data = f'mean: %.3f, standard deviation: %.3f'%(np.mean(hw['Mark']),np.std(hw['Mark']))
hist = sns.histplot(hw, x = 'Mark',kde=True)
hist.set_title('Implicit Differentiation Homework')
hist.annotate(data, (0.1,5.5))
plt.savefig('figures/p4_ch5_implicit.png',dpi=300)
```
